2. Решить задачу 1, распараллелив вычисления с помощью модуля `multiprocessing`. Для обработки каждого файла создать свой собственный процесс.

````python
#Программа не выполняется в юпитере

import multiprocessing as mp
from collections import Counter


def count_symbol(file_name):
    with open(f"{file_name}", 'r') as f:
        print(Counter(f.read().lower()))

# в юпитере это условие можно убрать
if __name__ == '__main__':
    output = mp.Queue()
    name = ['Dostoevskiy Fedor. Prestuplenie i nakazanie - BooksCafe.Net.txt',
            'Dostoevskiy Fedor. Igrok - BooksCafe.Net.txt']
    procs = [mp.Process(target=count_symbol, args=(name[i],)) for i in range(2)]

    for p in procs:
        p.start()

    for p in procs:
        p.join()
````

````python
import pandas as pd


def mean_for_tags(file_name):
    data = pd.read_csv(file_name, delimiter=';')
    new_data = data.groupby('id').mean(numeric_only=True)
    return new_data.to_dict()['n_steps']

def file_names(names):
    dictionary = {}
    for i in names:
        file_d = mean_for_tags(i)
        dictionary = {k: dictionary.get(k, 0) + file_d.get(k, 0) for k in set(dictionary) | set(file_d)}
    return dictionary


print(file_names(["id_tag_nsteps_0.csv", "id_tag_nsteps_1.csv", "id_tag_nsteps_2.csv", "id_tag_nsteps_3.csv",
                  "id_tag_nsteps_4.csv", "id_tag_nsteps_5.csv", "id_tag_nsteps_6.csv", "id_tag_nsteps_7.csv"]))
````

4. Решите задачу 3, распараллелив вычисления с помощью модуля `multiprocessing`. Для обработки каждого файла создайте свой собственный процесс. Определите, за какое время задача решается для всех файлов.

````python
import multiprocessing
from multiprocessing import Process
import pandas as pd


def mean_for_tags(file_name, return_list):
    data = pd.read_csv(file_name, delimiter=';')
    new_data = data.groupby('id').mean(numeric_only=True)
    return_list.append(new_data.to_dict()['n_steps'])


if __name__ == '__main__':

    manager = multiprocessing.Manager()
    return_list = manager.list()

    procs = [Process(target=mean_for_tags, args=(f"id_tag_nsteps_{i}.csv", return_list)) \
             for i in range(8)]

    for proc in procs:
        proc.start()

    for proc in procs:
        proc.join()

    dictionary = {}
    for file_d in return_list:
        dictionary = {k: dictionary.get(k, 0) + file_d.get(k, 0) for k in set(dictionary) | set(file_d)}
    print(dictionary)
    print("Олейников Александр")
````
