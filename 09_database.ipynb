1. Работая с базой данных `Chinook_Sqlite.sqlite`, найдите и выведите на экран имена и фамилии всех заказчиков из Канады

```python
con = sqlite3.connect("Chinook_Sqlite.sqlite")
cur = con.cursor()
```

```python
query = '''
SELECT firstname || ' ' || lastname
FROM Customer
WHERE country = 'Canada';
'''
```

```python
try:
    cur.execute(query)
except sqlite3.DatabaseError as err:
    print(f'Error: {err}')
else:
    for fullname in cur:
        print(*fullname)


print("-------------------\nАлександр Олейников")
```

```python
try:
    cur.execute(query2)
except sqlite3.OperationalError as err:
    print(f"Error: {err}")
else:
    for title in cur:
        print(*title)
    cur.close()
    con.close()


print("-------------------\nАлександр Олейников")
```

```python
con = sqlite3.connect('recipes.db')
cur = con.cursor()
```

2. Напишите и выполните SQL-запрос для создания таблицы рецептов `Recipe`. Таблица должна содержать следующие поля:
`id`, `name`, `minutes`, `submitted`, `description`, `n_ingredients`. Определитесь с типами и составом ключевых полей.

```python
query = '''
CREATE TABLE Recipes(
    id INT NOT NULL,
    name VARCHAR(100),
    minutes INT,
    submitted DATE,
    description VARCHAR(2000),
    n_ingredients INT,
    PRIMARY KEY(id))
'''
```

```python
try:
    cur.execute(query)
except sqlite3.OperationalError as err:
    print(f"Error: {err}")
else:
    print(f"Table created.")


print("-------------------\nАлександр Олейников")
```

3. Напишите и выполните SQL-запрос для создания таблицы отзывов Review. Таблица должна содержать следующие поля: id, user_id, recipe_id, date, rating, review. Определитесь с типами полей, набором ключевых полей. При помощи внешнего ключа соедините две таблицы.

```python
query2 = '''
CREATE TABLE Reviews(
    id INT NOT NULL,
    user_id INT,
    recipe_id INT,
    _date DATE,
    rating DECIMAL(1, 1),
    review VARCHAR(2000),
    PRIMARY KEY (id),
    FOREIGN KEY (recipe_id) REFERENCES Recipes(id))
'''
```

```python
try:
    cur.execute(query2)
except sqlite3.OperationalError as err:
    print(f"Error: {err}")
else:
    print('Table created.')


print("-------------------\nАлександр Олейников")
```

5. Найдите все рецепты, для выполнения которых нужно ровно 10 ингредиентов. Выведите на экран первые 5 из найденных рецептов.

```python
import sqlite3
con = sqlite3.connect("recipes.db")
cur = con.cursor()
```

```python
# можно и через LIMIT
query = '''
SELECT name, n_ingredients FROM Recipes
WHERE n_ingredients = 10;
'''
```

```python
try:
    cur.execute(query)
except sqlite3.OperationalError as err:
    print(f"Error {err}")
else:
    print(*cur.fetchmany(10), sep='\n') # метод который выводит 10 строк запроса и останавливается на 11


print("-------------------\nАлександр Олейников")
```

6. Найдите название рецепта, для выполнения которого требуется больше всего времени.

```python
query = '''
SELECT name, minutes FROM Recipes
WHERE minutes = (SELECT MAX(minutes) FROM Recipes)
'''
```

```python
try:
    cur.execute(query)
except sqlite3.OperationalError as err:
    print(f"Error: {err}")
else:
    for name, minutes in cur:
        print(f"{name} {minutes} минут")


print("-------------------\nАлександр Олейников")
```

8. Найдите кол-во отзывов с рейтингом 5.

```python
query = '''
SELECT Recipes.id
FROM Recipes JOIN Reviews ON Recipes.id=Reviews.recipe_id
GROUP BY Recipes.id
HAVING AVG(rating) = 5
'''
```

```python
try:
    cur.execute(query)
except sqlite3.OperationalError as err:
    print(f"Error: {err}")
else:
    print(len(cur.fetchall()))


print("-------------------\nАлександр Олейников")
```

9. Найдите кол-во уникальных рецептов, не имеющих отзывов с рейтингом, меньше 4.

```python
query = '''
SELECT DISTINCT Recipes.id
FROM Recipes JOIN Reviews ON Recipes.id=Reviews.recipe_id
GROUP BY Recipes.id
HAVING AVG(rating) >= 4
'''
```

```python
try:
    cur.execute(query)
except sqlite3.OperationalError as err:
    print(f"Error: {err}")
else:
    print(len(cur.fetchall()))


print("-------------------\nАлександр Олейников")
```

10. Найдите кол-во рецептов, опубликованных в 2010 году и имеющих длину не менее 15 минут.

11. Выберите id рецепта, название рецепта, id пользователя, оставившего отзыв, дату отзыва и рейтинг для тех рецептов, которые имеют не менее 3 ингредиентов. Отсортируйте результат по id рецепта.

```python
query = '''
SELECT rec.id, rec.name, rev.id, rev._date, rev.rating
FROM Recipes rec JOIN Reviews rev ON rec.id = rev.recipe_id
WHERE rec.n_ingredients >= 3
ORDER BY 1;
'''
```

```python
try:
    cur.execute(query)
except sqlite3.OperationalError as err:
    print(f"Error: {err}")
else:
    for line in cur:
        print(*line, sep=', ')


print("-------------------\nАлександр Олейников")
```
